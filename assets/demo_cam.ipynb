```python
from ultralytics import YOLO

# 1. Load the model 
# 'yolov8s.pt' is the 'small' version of YOLOv8
model = YOLO('yolov8s.pt') 

# 2. Run inference
# You can pass a URL, local path, PIL image, or numpy array
results = model.predict(source='https://ultralytics.com/images/bus.jpg', save=True, conf=0.25, imgsz=640)

# 3. Process results
for r in results:
    print(r.probs)  # print probabilities for classification
    r.show()        # display predictions on screen
    r.save(filename='result.jpg')  # save to disk
```
